# Estimating the Parameters with MAP
### from Chapter2: Estimating Probabilities, [Tom Mitchell's Machine Learning Book](www.cs.cmu.edu/~tom/mlbook)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# prior as beta function 
# same form as the likelihood
def beta(x, b1=0, b2=0):
    return x**b1 * (1-x)**b2
# prior params: the degree of certainty
gm0 = 42
gm1 = 18

# data(experiment)
alpha0 = 9
alpha1 = 6

test_x = np.linspace(0, 1, 1000)
prior = beta(test_x, gm0+1, gm1+1) 
likelihood = beta(test_x, alpha0, alpha1) 
posterior = likelihood * prior # Bayesian rule


gm0_l = 420 # larger gamma0 / gamma1: more certainty about prior
gm1_l = 180
prior_l = beta(test_x, gm0_l+1, gm1_l+1)
posterior_l = likelihood * prior_l

gm0_c = 32 # change gamma0 / gamma1: different prior
gm1_c = 28
prior_c = beta(test_x, gm0_c+1, gm1_c+1)
posterior_c = likelihood * prior_c

# plot
plt.figure()
plt.plot(test_x, prior/prior.sum(), label=r'$\beta$ prior, $\gamma_0=42, \gamma_1=18$')
plt.plot(test_x, posterior/posterior.sum(), label='posterior1')

plt.plot(test_x, prior_l/prior_l.sum(), label=r'$\beta$ prior, $\gamma_0=420, \gamma_1=180$')
plt.plot(test_x, posterior_l/posterior_l.sum(), label='posterior2')

plt.plot(test_x, prior_c/prior_c.sum(), label=r'$\beta$ prior, $\gamma_0=32, \gamma_1=28$')
plt.plot(test_x, posterior_c/posterior_c.sum(), label='posterior3')
plt.legend()

# \theta_MAP
theta_map = test_x[np.argmax(posterior)]
# \theta_MLE
theta_mle = test_x[np.argmax(likelihood)]

print(f"MAP parmas: {theta_map}")
print(f"MLE parmas: {theta_mle}")
```
